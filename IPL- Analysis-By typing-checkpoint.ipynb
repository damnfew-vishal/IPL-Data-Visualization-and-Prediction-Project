Now, we've to basically focus on the certain parameters like runs_left, balls_left and wickets_left. So, currently on first 2.

```python
delivery_df.groupby('match_id').cumsum()['total_runs_y']#cumsum(): It's basically the cummulative sum.
```

So as we can see it's basically giving us the cummulative sum or the total sum of the runs after each ball.
1, 1+0, 1+0, 1+2, 3+4 and so on.

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df
# So after this in the end it'll show one specific column mentioning the current score. And will further keeps on printing
# the cummulative sum of all the runs per ball.
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
# Here we are calculating the required runs thing. By subtracting the 'current score' from 'total_runs_x' 
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball']) #126 balls: Total balls.
# For example if we are considering 20 and 6 as over and ball respectively. So, after putting the value in the equation: it'll be
# 126 - 20*6 + 6 = 126 - 126 =0. This implies or this means that after this 20 overs and 6 balls. O balls are left further.
# For 19 and 1 //ly. It'll be 19*6+1=114+1=115. 126-115 so a total of 11 balls are left after that delivery.
```

The reason why it's going for 126 and not 120 is b/c in real it starts from 0.1 and 0.2 for 120 but here it's starting from 1.1 and 1.2 . 

```python
delivery_df
```

For wickets dismissed we don't have any direct source for that. But we can get help from the 'player_dismissed' column. If no one is out then 'NaN' is written.

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0") #Nan values are replaced with string of 0 values here.
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1") #Then here I'm looping through
#the column of this dataset and setting like if x's value is string of "0" then keep it as it is, if not then make it 1. 
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')#converted the whole column to integer
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values #Applied the same cummulative sum.
delivery_df['wickets'] = 10 - wickets #The subtracted those wickets from the 10. Which will give us the wickets left.
delivery_df.head()
```

```python
(delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
# crr = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
# (120 - balls left) and in order to convert it to no. of overs we divide it by 6 so rather than dividing directly we multipl-
# ied 6 with the current score.
```

```python
(delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
# runs_left/overs_left(same logic of no. of balls left divided by 6 but instead multiplied 6 to the numerator)
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df #will display the crr and rrr.
```

# One hot encoding

The first 3 columns of the data is in string format. So we'll be applying one hot encoding into that it is used.
In these cases, one-hot encoding comes in help because it transforms categorical data into numerical; in other words: it transforms strings into numbers so that we can apply our Machine Learning algorithms without any problems.

Column Transformer is a sciket-learn class used to create and apply separate transformers for numerical and categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder #imported the one hot encoder.

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
    #applied one hot encoding in the "batting_team", "bowling_team" and on the "city". B/c these 3 are in the forms of strings.
]
,remainder='passthrough')
```

So, we'll be importing the logistic regression along with the sklearn pipeline

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier 
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf), #transformer
    ('step2',LogisticRegression(solver='liblinear'))#Logistic regression
])
```

```python
pipe = Pipeline(steps=[
    ('step1',trf), #transformer
    ('step2',RandomForestClassifier())#This is by RandomForestClassification. This will basically gives us the accuracy of 99.9%
])
```

```python
X_train.describe()#will describe the current dataset
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

So above our logistic regression or RandomForestClassification regression is giving us the accuracy of 80% or 99.99% respectively. Depending on the case that I'm running of pipe = pipeline(steps of logistic regression and randomforestclassification.

```python
def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['total_runs_x']))
    
```

```python
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
    
```

WHAT THIS FUNCTION match_progression DO:
We give the whole data of this match that is of ball by ball detail, 2nd we talk about a particular match by match_id, and then 3rd give the model. 
Then what is does is after every over. It calculates the win probability of the chasing team and gives us in the format of dataframe.

```python
temp_df,target = match_progression(delivery_df,74,pipe)
temp_df
```

as we can see above with the match id of 1. We got each and every detail right from the end_of_over, runs_after_over, wickets_in_over and then lose and win. And since the target 207 is huge so we are actually getting the win and lose probability based on that only like if we lost the wicket the lose prob increased and win prob decreased. Similarly if the number of runs scored is less in a particular over then also the lose prob increased and win prob decreased. But, when the score increased compared to the last over there was a significant increase in the win probability and reduction in the lose prob.

```python
import matplotlib.pyplot as plt #plotting about the match.
plt.figure(figsize=(18,8))
plt.plot(temp_df['end_of_over'],temp_df['wickets_in_over'],color='yellow',linewidth=3)#Yellow: wickets in every over.
plt.plot(temp_df['end_of_over'],temp_df['win'],color='green',linewidth=4)#Green: winning probability
plt.plot(temp_df['end_of_over'],temp_df['lose'],color='red',linewidth=4)#Red: Loosing probability
plt.bar(temp_df['end_of_over'],temp_df['runs_after_over'])#Blue bars: runs after every over.
plt.title('Target-' + str(target))
```
